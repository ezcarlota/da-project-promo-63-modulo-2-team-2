## Extracción de canciones en JSON

```python
# usamos la forma json para acceder a la base de datos

with open('definitivo.json') as archivo:
    data = json.load(archivo)
```

```python
canciones_pop_2010_2015 = []

anios = [2013, 2014, 2015]
genres = ['pop']

for cancion in data:
    if cancion['genre'] in genres:
        if cancion['year'] in anios:
            canciones_pop_2010_2015.append(cancion)

canciones_pop_2010_2015
```

```python
artistas_set = set()

for canciones in canciones_pop_2010_2015:
    nombre = canciones.get("artist_name")
    if nombre:
        artistas_set.add(nombre.strip())

artistas = sorted(artistas_set)
print("Artistas únicos:", len(artistas))
```

## Extracción de información de Artistas en Last FM

```python
from dotenv import load_dotenv
load_dotenv()

api_key = os.getenv("api_key") # accedemos a la variable api_key que está en el archivo .env
```

```python
# buscar el endpoint que nos interesa --> 
# codigo de prueba con 1 solo artista

datos_artista = []

for artista in ['Adele']:
    url = "https://ws.audioscrobbler.com/2.0/"

    llamada = requests.get(url, params={'api_key': api_key,
                                        'method': 'artist.getinfo',
                                        'artist': {artista},
                                        'format': 'json'})
    datos = llamada.json()
    datos_artista.append(datos)
```

```python
for elemento in datos_artista:
    print(elemento['artist'].keys())
```

```python
import re
```

```python
def limpiar_summary(texto):
    if not texto:
        return None

    # Eliminar etiquetas HTML
    texto = re.sub(r"<.*?>", "", texto)

    # Eliminar saltos de línea
    texto = texto.replace("\n", " ")

    # Eliminar espacios múltiples
    texto = re.sub(r"\s+", " ", texto)

    # Eliminar coletilla típica de Last.fm
    texto = texto.replace("Read more on Last.fm", "")

    return texto.strip()
```

```python
info_limpia = []

for elemento in datos_artista:

    if "artist" not in elemento:
        continue

    artista = elemento["artist"]

    similares = artista.get("similar", {}).get("artist", [])
    nombres_similares = [a.get("name") for a in similares]

    resumen_limpio = limpiar_summary(
        artista.get("bio", {}).get("summary")
    )

    info_limpia.append({
        "artist_name": artista.get("name"),
        "bio_summary": resumen_limpio,
        "listeners": artista.get("stats", {}).get("listeners"),
        "playcount": artista.get("stats", {}).get("playcount"),
        "similar_artists": nombres_similares
    })
```

## Ahora con la lista entera de artistas

```python
import time
import requests

def obtener_artista(nombre, api_key, reintentos=2, timeout=10):

    for intento in range(reintentos + 1):
        try:
            response = requests.get(
                "https://ws.audioscrobbler.com/2.0/",
                headers={"User-Agent": "adalab-project"},
                params={
                    "method": "artist.getinfo",
                    "artist": nombre,
                    "api_key": api_key,
                    "format": "json"
                },
                timeout=timeout
            )

            if response.status_code == 200:
                return response.json()

        except requests.exceptions.Timeout:
            if intento < reintentos:
                time.sleep(1)  # backoff
            else:
                print(f"Timeout definitivo con {nombre}")
                return None

        except requests.exceptions.RequestException as e:
            print(f"Error con {nombre}: {e}")
            return None
```

```python
info_limpia = []

for artista_nombre in artistas_set:

    print(f"Procesando: {artista_nombre}")

    data = obtener_artista(artista_nombre, api_key)

    if not data or "artist" not in data:
        continue

    artista = data["artist"]

    similares = artista.get("similar", {}).get("artist", [])
    nombres_similares = [a.get("name") for a in similares]

    resumen_limpio = limpiar_summary(
        artista.get("bio", {}).get("summary")
    )

    info_limpia.append({
        "artist_name": artista.get("name"),
        "bio_summary": resumen_limpio,
        "listeners": artista.get("stats", {}).get("listeners"),
        "playcount": artista.get("stats", {}).get("playcount"),
        "similar_artists": nombres_similares
    })

    time.sleep(0.2)
```
