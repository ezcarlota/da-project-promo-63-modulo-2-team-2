```python
datos_artista = []

for artista in artistas_unicos:
    url = "https://ws.audioscrobbler.com/2.0"

    llamada = requests.get(url, params={'api_key': api_key,
                                        'method': 'artist.getinfo',
                                        'artist': {artista},
                                        'format': 'json'})
    datos = llamada.json()
    datos_artista.append(datos)
```

```python
info_limpia = []
```

```python
for elemento in datos_artista:
    
    if "artist" not in elemento:
        continue
    
    artista = elemento["artist"]
    
    info_limpia.append({
        "artist_name": artista.get("name"),
        "bio_summary": artista.get("bio", {}).get("summary"),
        "listeners": artista.get("stats", {}).get("listeners"),
        "playcount": artista.get("stats", {}).get("playcount"),
        "similar_artists": artista.get("similar", {}).get("artist")
    })
```

```python
info_limpia
```

```python
for artista in info_limpia:
    # extraer solo los nombres de los similares
    nombres_similares = [s['name'] for s in artista.get('similar_artists', [])]
    
    # reemplazar la lista original por solo los nombres
    artista['similar_artists'] = nombres_similares
```

```python
info_limpia
```

```python
for artista in info_limpia:
    bio = artista.get("bio_summary")

    if bio:
        bio = bio.split("<a href=")[0]
        bio = bio.replace("Read more on Last.fm", "")
        bio = bio.replace("\n", " ")
        bio = " ".join(bio.split())

        if not bio.strip():
            bio = None

    artista["bio_summary"] = bio
```

```python
info_limpia
```

```python
for artista in info_limpia:
    artista["listeners"] = int(artista["listeners"]) if artista["listeners"] else 0
    artista["playcount"] = int(artista["playcount"]) if artista["playcount"] else 0
```

```python
df_similares = []
```

```python
similar_data = []
```

```python
for artista in info_limpia:
    artista_principal = artista["artist_name"]
    # si quieres solo 5 similares
    for pos, similar in enumerate(artista["similar_artists"][:5], start=1):
        similar_data.append(
            {
                "artist_name": artista_principal,
                "similar_artist_name": similar,
                "rank": pos,
            }
        )
```

```python
df_similares = pd.DataFrame(similar_data)
df_similares = df_similares.drop_duplicates().reset_index(drop=True)
```
