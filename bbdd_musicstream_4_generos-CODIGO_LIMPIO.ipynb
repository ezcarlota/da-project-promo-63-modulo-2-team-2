```python
try:
    # nos conectamos al servidor de mysql, no especificamos schema porque vamos a crearlo
    
    cnx = mysql.connector.connect(
        user='root',
        password= input('password_MySQL'),
        host='127.0.0.1'
        # auth_plugin = 'mysql_native_password'
    )
    print("¡Conexión exitosa a MySQL!") #
except mysql.connector.Error as err:
    # Si es un error de acceso denegado (ej. contraseña o usuario incorrecto)
    if err.errno == errorcode.ER_ACCESS_DENIED_ERROR:
        print("Algo está mal con tu nombre de usuario o contraseña.") #
    # Si la base de datos no existe
    elif err.errno == errorcode.ER_BAD_DB_ERROR:
        print("La base de datos no existe.") #
    # Para cualquier otro tipo de error
    else:
        print(err) #
        print("Código de Error:", err.errno) #
        print("SQLSTATE", err.sqlstate) #
        print("Mensaje", err.msg) #
```

```python
mycursor = cnx.cursor() # Iniciamos el cursor
mycursor.execute("CREATE SCHEMA IF NOT EXISTS musicstream;") # creamos schema
mycursor.execute("USE musicstream") # nos movemos al schema que acabamos de crear
# abrimos el cursor asociado a esta conexión
```

```python
mycursor = cnx.cursor() 
# abrimos el cursor asociado a esta conexión
```

```python
## Tabla artistas

tabla_artista = """CREATE TABLE IF NOT EXISTS artistas (
  id_artista INT NOT NULL AUTO_INCREMENT,
  nombre_artista VARCHAR(100) NULL,
  popularidad INT NULL,
  reproducciones BIGINT NULL,
  bio LONGTEXT NULL,
  PRIMARY KEY (id_artista),
  UNIQUE INDEX id_artista_UNIQUE (id_artista ASC) VISIBLE)
ENGINE = InnoDB;"""

mycursor.execute(tabla_artista)
cnx.commit()
print("Tabla creada correctamente")
```

```python
## Tabla genero

tabla_genero = """CREATE TABLE IF NOT EXISTS genero (
        id_genero INT NOT NULL AUTO_INCREMENT,
        nombre_genero VARCHAR(45) NOT NULL,
        PRIMARY KEY (id_genero),
        UNIQUE INDEX nombre_genero_UNIQUE (nombre_genero ASC) VISIBLE)
        ENGINE = InnoDB;"""

mycursor.execute(tabla_genero)
cnx.commit()
print("Tabla creada correctamente")
```

```python
## Table canciones

tabla_canciones = """CREATE TABLE IF NOT EXISTS canciones (
        id_cancion INT NOT NULL AUTO_INCREMENT,
        nombre_cancion VARCHAR(250) NOT NULL,
        año_lanzamiento YEAR(4) NULL,
        id_genero INT NOT NULL,
        PRIMARY KEY (id_cancion),
        INDEX fk_canciones_genero_idx (id_genero ASC) VISIBLE,
        CONSTRAINT fk_canciones_genero
        FOREIGN KEY (id_genero)
        REFERENCES `musicstream`.`genero` (id_genero)
        ON DELETE NO ACTION
        ON UPDATE NO ACTION)
        ENGINE = InnoDB;"""

mycursor.execute(tabla_canciones)
cnx.commit()
print("Tabla creada correctamente")
```

```python
## Tabla artistas similares

tabla_artistas_similares = """CREATE TABLE IF NOT EXISTS artistas_similares (
        id_artista_similar INT NOT NULL AUTO_INCREMENT,
        nombre_artistas_similar VARCHAR(250) NULL,
        PRIMARY KEY (`id_artista_similar`))
        ENGINE = InnoDB;"""

mycursor.execute(tabla_artistas_similares)
cnx.commit()
print("Tabla creada correctamente")
```

```python
## Tabla FK_artistas_artistas_similares

tabla_FK_artistas_artistas_similares = """CREATE TABLE IF NOT EXISTS FK_artistas_artistas_similares (
        id_artista_similar INT NOT NULL,
        id_artista INT NOT NULL,
        PRIMARY KEY (`id_artista_similar`, `id_artista`),
        INDEX `fk_artistas similares_has_artistas_artistas1_idx` (`id_artista` ASC) VISIBLE,
        INDEX `fk_artistas similares_has_artistas_artistas similares_idx` (`id_artista_similar` ASC) VISIBLE,
        CONSTRAINT `fk_artistas similares_has_artistas_artistas similares`
          FOREIGN KEY (`id_artista_similar`)
          REFERENCES artistas_similares (`id_artista_similar`)
          ON DELETE NO ACTION
          ON UPDATE NO ACTION,
        CONSTRAINT `fk_artistas similares_has_artistas_artistas1`
          FOREIGN KEY (`id_artista`)
          REFERENCES artistas (`id_artista`)
          ON DELETE NO ACTION
          ON UPDATE NO ACTION)
      ENGINE = InnoDB;"""

mycursor.execute(tabla_FK_artistas_artistas_similares)
cnx.commit()
print("Tabla creada correctamente")
```

```python
tabla_fk_canciones_artistas = """CREATE TABLE IF NOT EXISTS fk_artistas_canciones (
  id_artista INT NOT NULL,
  id_cancion INT NOT NULL,
  PRIMARY KEY (id_artista, id_cancion),
  INDEX fk_artistas_has_canciones_canciones1_idx (id_cancion ASC) VISIBLE,
  INDEX fk_artistas_has_canciones_artistas1_idx (id_artista ASC) VISIBLE,
  CONSTRAINT fk_artistas_has_canciones_artistas1
    FOREIGN KEY (id_artista)
    REFERENCES artistas (id_artista)
    ON DELETE NO ACTION
    ON UPDATE NO ACTION,
  CONSTRAINT fk_artistas_has_canciones_canciones1
    FOREIGN KEY (id_cancion)
    REFERENCES canciones (id_cancion)
    ON DELETE NO ACTION
    ON UPDATE NO ACTION)
ENGINE = InnoDB;"""

mycursor.execute(tabla_fk_canciones_artistas)
cnx.commit()
print("Tabla creada correctamente")
```

# INSERTAMOS DATOS REGGAETON

```python
## insercion de datos artistas reggaeton


import csv

insertados = 0
ignorados = 0
errores = 0

with open("datos_reggaeton_last.csv", newline="", encoding="utf-8") as csvfile:
    reader = csv.DictReader(csvfile)
    print("Columnas CSV:", reader.fieldnames)

    for row in reader:
        try:
            nombre = (row.get("artist_name") or "").strip()
            bio = row.get("bio_summary") or None

            # listeners -> popularidad (INT o NULL)
            listeners = (row.get("listeners") or "").strip()
            if listeners == "":
                popularidad = None
            else:
                popularidad = int(float(listeners))

            # playcount -> reproducciones (INT o NULL)
            playcount = (row.get("playcount") or "").strip()
            if playcount == "":
                reproducciones = None
            else:
                reproducciones = int(float(playcount))

            #  INSERT IGNORE evita duplicados 
            mycursor.execute(
                """
                INSERT IGNORE INTO artistas (nombre_artista, popularidad, reproducciones, bio)
                VALUES (%s, %s, %s, %s)
            """,
                (nombre, popularidad, reproducciones, bio),
            )

            if mycursor.rowcount == 1:
                insertados += 1
            else:
                ignorados += 1

        except Exception as e:
            errores += 1
            print("Error con fila:", row)
            print("   ", e)

cnx.commit()

print("Insertados:", insertados)
print("Ignorados (duplicados):", ignorados)
print("Errores:", errores)
```

```python
## insercion de datos canciones reggaeton


import csv
import mysql.connector

# ---------------- CONFIGURACIÓN ----------------

ARCHIVO_CSV_REGGAETON = "datos_reggaeton_spotify.csv"
ID_GENERO_REGGAETON = 1


mycursor = cnx.cursor(buffered=True)

# ---------------- CONTADORES ----------------

canciones_insertadas = 0
relaciones_insertadas = 0
artistas_no_encontrados = 0
errores = 0

# ---------------- ETL ----------------

with open(ARCHIVO_CSV_REGGAETON, newline="", encoding="utf-8") as archivo:
    reader = csv.DictReader(archivo)

    for fila in reader:
        try:
            # -------- TRANSFORM --------
            artista = fila["artist_name"].strip()
            cancion = fila["track_name"].strip()
            anio = fila["year"].strip()

            if not artista or not cancion:
                continue

            anio = int(anio) if anio.isdigit() else None

            # -------- LOAD: ARTISTA --------
            mycursor.execute(
                "SELECT id_artista FROM artistas WHERE nombre_artista = %s",
                (artista,)
            )
            resultado = mycursor.fetchone()

            if not resultado:
                artistas_no_encontrados += 1
                continue

            id_artista = resultado[0]

            # -------- LOAD: CANCIÓN --------
            mycursor.execute(
                """
                INSERT INTO canciones (nombre_cancion, año_lanzamiento, id_genero)
                VALUES (%s, %s, %s)
                ON DUPLICATE KEY UPDATE id_cancion = LAST_INSERT_ID(id_cancion)
                """,
                (cancion, anio, ID_GENERO_REGGAETON)
            )

            id_cancion = mycursor.lastrowid
            canciones_insertadas += 1

            # -------- LOAD: RELACIÓN --------
            mycursor.execute(
                """
                INSERT IGNORE INTO fk_artistas_canciones (id_artista, id_cancion)
                VALUES (%s, %s)
                """,
                (id_artista, id_cancion)
            )

            if mycursor.rowcount == 1:
                relaciones_insertadas += 1

        except Exception as e:
            errores += 1
            print("Error en fila:", fila)
            print("Detalle:", e)

# ---------------- COMMIT ----------------

cnx.commit()

# ---------------- RESULTADOS ----------------

print("\nProceso ETL terminado")
print("Canciones procesadas:", canciones_insertadas)
print("Relaciones creadas:", relaciones_insertadas)
print("Artistas no encontrados:", artistas_no_encontrados)
print("Errores:", errores)
```

```python
## insercion de datos canciones rock

import csv
import mysql.connector

# ---------------- CONFIGURACIÓN ----------------

ARCHIVO_CSV_ROCK = "artistas_rock_spotify_2013-2015.csv"
ID_GENERO_ROCK = 5


mycursor = cnx.cursor(buffered=True)

# ---------------- CONTADORES ----------------

canciones_insertadas = 0
relaciones_insertadas = 0
artistas_no_encontrados = 0
errores = 0

# ---------------- ETL ----------------

with open(ARCHIVO_CSV_ROCK, newline="", encoding="utf-8") as archivo:
    reader = csv.DictReader(archivo)

    for fila in reader:
        try:
            # -------- TRANSFORM --------
            artista = fila["artist_name"].strip()
            cancion = fila["track_name"].strip()
            anio = fila["year"].strip()

            if not artista or not cancion:
                continue

            anio = int(anio) if anio.isdigit() else None

            # -------- LOAD: ARTISTA --------
            mycursor.execute(
                "SELECT id_artista FROM artistas WHERE nombre_artista = %s",
                (artista,)
            )
            resultado = mycursor.fetchone()

            if not resultado:
                artistas_no_encontrados += 1
                continue

            id_artista = resultado[0]

            # -------- LOAD: CANCIÓN --------
            mycursor.execute(
                """
                INSERT INTO canciones (nombre_cancion, año_lanzamiento, id_genero)
                VALUES (%s, %s, %s)
                ON DUPLICATE KEY UPDATE id_cancion = LAST_INSERT_ID(id_cancion)
                """,
                (cancion, anio, ID_GENERO_ROCK)
            )

            id_cancion = mycursor.lastrowid
            canciones_insertadas += 1

            # -------- LOAD: RELACIÓN --------
            mycursor.execute(
                """
                INSERT IGNORE INTO fk_artistas_canciones (id_artista, id_cancion)
                VALUES (%s, %s)
                """,
                (id_artista, id_cancion)
            )

            if mycursor.rowcount == 1:
                relaciones_insertadas += 1

        except Exception as e:
            errores += 1
            print("Error en fila:", fila)
            print("Detalle:", e)

# ---------------- COMMIT ----------------

cnx.commit()

# ---------------- RESULTADOS ----------------

print("\nProceso ETL terminado")
print("Canciones procesadas:", canciones_insertadas)
print("Relaciones creadas:", relaciones_insertadas)
print("Artistas no encontrados:", artistas_no_encontrados)
print("Errores:", errores)
```

# Inserción datos Artistas Similares

```python
import csv

# Cargar artistas principales en memoria
mycursor.execute("SELECT id_artista, nombre_artista FROM artistas")
artistas_principales = {
    nombre.strip(): id_artista
    for id_artista, nombre in mycursor.fetchall()
}

with open("artista_similar_pop.csv", newline="", encoding="utf-8") as archivo:
    lector = csv.DictReader(archivo)

    for fila in lector:
        artista_principal = (fila.get("artist_name") or "").strip()
        artista_similar = (fila.get("similar_artist_name") or "").strip()

        if not artista_principal or not artista_similar:
            continue

        if artista_principal not in artistas_principales:
            continue

        id_artista = artistas_principales[artista_principal]

        # -------- resolver id_artista_similar --------

        if artista_similar in artistas_principales:
            id_similar = artistas_principales[artista_similar]

        else:
            mycursor.execute(
                """
                SELECT id_artista_similar
                FROM artistas_similares
                WHERE nombre_artistas_similar = %s
                """,
                (artista_similar,)
            )
            resultado = mycursor.fetchone()

            if resultado:
                id_similar = resultado[0]
            else:
                mycursor.execute(
                    """
                    INSERT INTO artistas_similares (nombre_artistas_similar)
                    VALUES (%s)
                    """,
                    (artista_similar,)
                )
                id_similar = mycursor.lastrowid

        # -------- insertar relación SIEMPRE --------

        mycursor.execute(
            """
            INSERT IGNORE INTO fk_artistas_artistas_similares
            (id_artista, id_artista_similar)
            VALUES (%s, %s)
            """,
            (id_artista, id_similar)
        )

cnx.commit()
```
